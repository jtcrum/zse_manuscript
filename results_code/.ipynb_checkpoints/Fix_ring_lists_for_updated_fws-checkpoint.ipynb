```python
def get_fw_data(file,code):
    file = open(file,'r')
    data = file.readlines()
    file.close()
    start = False
    text = []
    for i,line in enumerate(data):
        fields = line.split()
        if fields[0] == code:
            start = i
        if start:
            text.append(line)
        if start and fields[0]== '******************************':
            end = i
            break

    file.close()
    return start,end,text
s,e,text = get_fw_data('crum_rings.txt','FRA')
print(text[0])
print(text[-1])
print(text[-2])
```

```python
fws = ['ITH','IWW','SSY','CFI','NAB','CZP','GON','MTT','NPO','PAR','RWY','SFS','TUN']
for f in fws:
    s,e,text = get_fw_data('updated_crum_rings.txt',f)
    s2,e2,t2 = get_fw_data('crum_rings.txt',f)
    with open('crum_rings.txt','r+') as file:
        d = file.readlines()
        file.seek(0)
        for i,line in enumerate(d):
            if i in range(s2,e2+1):
                print(line)
    break
```

```python
file = open('goetzke_rings.txt',"a")
file.write('****************************** \n')
file.close()
fws = ['FRA','ITH','IWW','SSY','CFI','NAB','CZP','GON','MTT','NPO','PAR','RWY','SFS','TUN']
for f in fws:
    file = open('updated_goetzke_rings.txt',"a")
    file.write('Framework \n')
    file.write(f + '\n')
    file.write('------------------------------ \n')
    flag = True
    # O-site rings
    file.write('O_rings \n')
    file.write('------------------------------ \n')
    z = framework(f)
    repeat = atoms_to_graph(z,0,40)[2]
    z2 = z.repeat(repeat)
    index_list = np.arange(len(z2)).reshape(np.prod(repeat),len(z))
    labels = site_labels(z,f)
    os,om,oinds = get_osites(f)
    for i,o in enumerate(os):
        try:
            c,r,ra,a = get_rings(z,oinds[i],max_ring=20)
            ring_list = [str(x) for x in c]
            ring_list = '_'.join(ring_list)
            file.write('{0}:{1} \n'.format(o,ring_list))
            file.write('{0} Ring Compositions \n'.format(o))
            for p in r:
                temp = []
                for x in p:
                    ind = int(np.where(index_list==x)[1])
                    temp.append(labels[ind])
                temp = '_'.join(temp)
                file.write('{0}-MR | {1} \n'.format(int(len(p)/2),temp))
            file.write('------------------------------ \n')
        except:
            file.write('{0}:FAILED \n'.format(o))
    file.write('T_rings \n')
    file.write('------------------------------ \n')
    ts,tm,tinds = get_tsites(f)
    for i,t in enumerate(ts):
        try:
            c,r,ra,a = get_rings(z,tinds[i],max_ring=20)
            ring_list = [str(x) for x in c]
            ring_list = '_'.join(ring_list)
            file.write('{0}:{1} \n'.format(t,ring_list))
            file.write('{0} Ring Compositions \n'.format(t))
            for p in r:
                temp = []
                for x in p:
                    ind = int(np.where(index_list==x)[1])
                    temp.append(labels[ind])
                temp = '_'.join(temp)
                file.write('{0}-MR | {1} \n'.format(int(len(p)/2),temp))
            file.write('------------------------------ \n')
        except:
            file.write('{0}:FAILED \n'.format(t))
    file.write('****************************** \n')
    file.close()
```
